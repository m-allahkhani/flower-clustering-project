```python
# from google.colab import drive
# drive.mount('/content/drive')  # use this if you are using google colab
import pickle
from sklearn.cluster import KMeans
import random
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import math
```

```python
def load_dataset(name_file):
 # Replace with your desired directory path
    with open(name_file, 'rb') as f:
        data_dict = pickle.load(f)
    return data_dict
```

```python
def classify_kmeans(train_data, train_label, numberCluster):
    train_data = np.array(train_data)
    scaler = StandardScaler()
    normalized_data = scaler.fit_transform(train_data[:, :])
    kmeans = KMeans(n_clusters=numberCluster)
    kmeans.fit(normalized_data)
    
    # Get cluster assignments for each data point
    cluster_labels = kmeans.predict(normalized_data)
    
    # Initialize lists to store clusters and labels in each cluster
    clusters = [[] for _ in range(numberCluster)]
    labels_in_clusters = [[] for _ in range(numberCluster)]
    
    # Assign each data point to its corresponding cluster
    for i in range(len(train_data)):
        cluster_id = cluster_labels[i]
        clusters[cluster_id].append(train_data[i])
        labels_in_clusters[cluster_id].append(train_label[i])
    
    # Get the center labels (labels of the cluster centers)
    center_labels = kmeans.predict(kmeans.cluster_centers_)

    return kmeans.cluster_centers_, clusters, center_labels, labels_in_clusters
```

```python
def calculate_distance(centroid_data, test_data):
    
    distance_cols = []
    for j in range(len(test_data)):
        distance_cols.append(math.sqrt((centroid_data[j]-test_data[j])**2))
    result = sum(distance_cols) / len(distance_cols)

    return result
```

```python
def knn_clusters(k, cluster_centers_, centers_labels, test_data):
    model = KNeighborsClassifier(n_neighbors=k, metric=calculate_distance , algorithm='auto')
    model.fit(cluster_centers_, centers_labels)
    # Predict the class labels for the test set
    predictions = model.predict(test_data)
    # print(len(predictions))
    print(predictions)
    return predictions
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

def knn_dataset(k, x_train, y_train, test_data):
    model = KNeighborsClassifier(n_neighbors=k, metric=calculate_distance, algorithm='auto')
    
    # Convert x_train and y_train to NumPy arrays
    x_train = np.array(x_train)
    y_train = np.array(y_train)
    
    # Reshape test_data to ensure it's a 2D array
    test_data = np.array(test_data).reshape(1, -1)
    
    model.fit(x_train, y_train)

    prediction = model.predict(test_data)
    return prediction
```

```python
def calculate_accuracy(predictions,y_test):
    result = 0
    for i in range(len(predictions)):
        if predictions[i]== y_test[i]:
            result += 1
        
    result /= len(predictions)
    return result
```

```python
def calculate_score(accuracy,k,p = -0.2):
    return k*p + accuracy
```
